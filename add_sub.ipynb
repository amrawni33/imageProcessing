```python

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("download.png", cv2.IMREAD_GRAYSCALE)

def add(img):

    plt.imshow(img, cmap="gray")
    plt.title("Original Image")
    plt.show()

    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if img[x][y] + 127 > 255:
                img[x][y] = 255
            else:
                img[x][y] += 127

    plt.imshow(img, cmap="gray")
    plt.title("Image After Adding 127 to Each Pixel")
    plt.show()

add(img)
```

```python

def subtract(img):

    plt.imshow(img, cmap="gray")
    plt.title("Original Image")
    plt.show()

    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if img[x][y] < 127:
                img[x][y] = 0
            else:
                img[x][y] -= 127

    plt.imshow(img, cmap="gray")
    plt.title("Image After subtracting 127 from Each Pixel")
    plt.show()

subtract(img)
```
