Log 

```python
import math
import cv2 
import matplotlib.pyplot as plt

img = cv2.imread("maxresdefault.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap="gray")
plt.title("original image")
plt.show()

def logFun(img):
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            img[x][y] = math.log2(1+img[x][y])
    plt.imshow(img, cmap="gray")
    plt.title("image after processing")
    plt.show()

logFun(img)
```

```python
import math
import cv2 
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("maxresdefault.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap="gray")
plt.title("original image")
plt.show()

def logExp(img, gamma):
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            img[x][y] = np.power(img[x][y] / 255.0, gamma) * 255.0
    plt.imshow(img, cmap="gray")
    plt.title("image after processing")
    plt.show()

logExp(img, 2)
```
