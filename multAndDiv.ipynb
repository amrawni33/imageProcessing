# multiplication and division 


```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('Desktop/me.png')

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap="gray")
plt.title("Before")

# Create a copy of the image to perform the operations on
new_img = np.copy(img)

for x in range(new_img.shape[0]):
    for y in range(new_img.shape[1]):
        new_pixel_value = new_img[x][y] * 2
        new_img[x][y] = np.minimum(new_pixel_value, 255)

plt.subplot(1, 2, 2)
plt.imshow(new_img, cmap="gray")
plt.title("After")

plt.show()
```
